```python
# Review data with empty tags

empty_tags_mask = df_tags_removed['tags'].str.len() == 0
df_with_empty_tags = df_tags_removed[empty_tags_mask]

print(len(df_with_empty_tags))
df_with_empty_tags.head(40)
```

```python
# Remove data with empty tags

df_with_no_empty_tags = df_tags_removed[~empty_tags_mask]

print(len(df_with_no_empty_tags))
df_with_no_empty_tags.head(30)
```

```python
import numpy as np

# Add additional column to distinguish meal, side-dishes, and desserts
df_with_meal_types = df_with_no_empty_tags.copy()

condition = [
    df_with_meal_types['tags'].apply(lambda x: any(tag in x for tag in ['desserts', 'pies-and-tarts', 'cakes', 'cookies-and-brownies', 'chocolates', 'drop-cookies'])),
    df_with_meal_types['tags'].apply(lambda x: any(tag in x for tag in ['side-dishes', 'appetizers', 'condiments-etc', 'dips', 'sauces']))
]

options = [
    'dessert',
    'side-dish'
]

df_with_meal_types['meal_type'] = np.select(condition, options, default='meal')

print(len(df_with_meal_types))
df_with_meal_types.head(30)
```

```python
# Cleaning tags as the data set now has a separate column to categorize the meal type

additional_tags_to_remove = [
    'side-dish',
    'main-dish',
    'dessert',
    'cookie',
    'pie',
    'tart',
    'cake',
    'chocolate',
    'appetizer',
    'condiment',
    'dip',
    'sauce',
    'beverage',
    'cocktail'
]

df_additional_tag_removed = df_with_meal_types.copy()

df_additional_tag_removed['tags'] = df_additional_tag_removed['tags'].apply(
    lambda tag_list: [
        tag for tag in tag_list 
        if tag.strip() and not any(word in tag for word in additional_tags_to_remove)
    ]
)

df_additional_tag_removed = df_additional_tag_removed[df_additional_tag_removed['tags'].str.len() > 0]

df_additional_tag_removed.head(20)
```

```python
# Add one more column 'food type' to distinguish dietary restriction
meat_keywords = ['meat', 'chicken', 'beef', 'pork', 'lamb', 'turkey', 'sausage', 'bacon', 'ham']
seafood_keywords = ['fish', 'tuna', 'salmon', 'shrimp', 'crab', 'squid', 'seafood', 'octopus']
vegan = ['vegan']
vegetarian = ['vegetarian']

conditions = [
    df_additional_tag_removed['tags'].apply(lambda x: any(tag in x for tag in meat_keywords)) | 
    df_additional_tag_removed['name'].str.lower().apply(lambda name: any(word in name for word in meat_keywords)),
    df_additional_tag_removed['tags'].apply(lambda x: (any(tag in x for tag in seafood_keywords)) & 
                                            (not any(tag in x for tag in meat_keywords))),
    df_additional_tag_removed['tags'].apply(lambda x: any(tag in x for tag in vegan)),
    df_additional_tag_removed['tags'].apply(lambda x: any(tag in x for tag in vegetarian))
]

options = [
    'meat-base',
    'pescetarian',
    'vegan',
    'vegetarian'
]

df_food_type_added = df_additional_tag_removed.copy()
df_food_type_added['food_type'] = np.select(conditions, options, default='uncategorized')

df_food_type_added.head(30)
```

```python
# More cleaning

more_tags_to_remove = [
    'meat',
    'seafood',
    'vegan',
    'vegetarian'
]

df_more_tag_removed = df_food_type_added.copy()

df_more_tag_removed['tags'] = df_food_type_added['tags'].apply(
    lambda tag_list: [tag for tag in tag_list if tag not in more_tags_to_remove]
)

df_more_tag_removed = df_more_tag_removed[df_more_tag_removed['tags'].str.len() > 0]

print(len(df_more_tag_removed)/len(df)) # Percentage cleaned off, roughly ~53.3% of data are removed but still have more than 100,000 recipes
df_more_tag_removed.head(20)
```

```python
import threading
from queue import Queue
import requests
import time

prompt = f"""
You are a recipe categorization assistant. 
As you can see, the 'food_type' category is currently 'uncategorized'.
Your task is to update "food_type" from 'uncategorized' into one the followings based on their "name", 'tags', "ingredients", and "description". 
Category options: 1) vegetarian 2) vegan 3) meat-base 4) pescetarian.

Each recipe object in the array must contain all the original keys and you will only update the "food_type."
If you can't categorize, leave it as 'uncategorized'. 

Recipes to categorize:
"""

NUM_THREADS = 5
max_retries = 5
batch_queue = Queue()
updated_recipes = []

result_lock = threading.Lock()
progress_lock = threading.Lock()
processed_count = 0

api_url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash-preview-05-20:generateContent?key={API_KEY}"

def worker_thread():
    global processed_count
    
    while True:
        try:
            retries = 0
            batch = batch_queue.get(block=False)
            batch_json = json.dumps(batch.to_dict('records'))
            
            full_prompt = f"{prompt}\n{batch_json}"
            payload = {
                "contents": [
                    {
                        "parts": [{"text": full_prompt}]
                    }
                ],
                "generationConfig": {
                    "responseMimeType": "application/json"
                }
            }
            
            while retries < max_retries:
                try: 
                    print(f"Sending requests for batch (current queue size: {batch_queue.qsize()})... (Attempt {retries + 1})")
                    response = requests.post(api_url, json=payload)
                    response.raise_for_status()
                    
                    result = response.json()
                    
                    if result.get("candidates"):
                        json_response_text = result["candidates"][0]["content"]["parts"][0]["text"]
                        llm_response = json.loads(json_response_text)
                        
                        with result_lock:
                            updated_recipes.extend(llm_response)
                            
                        with progress_lock:
                            processed_count += 1
                            print(f"Successfully processed batch {processed_count}")
                            
                            break
                        
                    else:
                        print(f"API response missing candidates. Skipping.")
                        break
                
                except requests.exceptions.HTTPError as e:
                    if e.response.status_code in [429, 503, 500, 502, 504, 408]:
                        delay = 2**retries
                        print(f"Transient HTTP Error {e.response.status_code}: {e.response.text}. Retrying in {delay}s.")
                        time.sleep(delay)
                        retries += 1
                    else:
                        print(f"Permanent HTTP Error {e.response.status_code}: {e.response.text}. Skipping.")
                        break
                
                except (KeyError, IndexError, json.JSONDecodeError) as e:
                    print(f"Failed to parse API response: {e}. Skipping.")
                    break
                
                except requests.exceptions.RequestException as e:
                    print(f"An unexpected error occurred during the API call: {e}. Retrying.")
                    delay = 2**retries
                    time.sleep(delay)
                    retries += 1
            if retries == max_retries:
                print(f"Max retries exceeded.")
                
            batch_queue.task_done()
        except Queue.empty:
            break
        
def main():
    """
    Main function to orchestrate the multithreaded processing.
    """
    print(f"Starting multithreaded processing with {NUM_THREADS} threads...")

    # Populate the queue with all the batches.
    for batch in batches:
        batch_queue.put(batch)

    # Create and start the worker threads.
    threads = []
    for _ in range(NUM_THREADS):
        thread = threading.Thread(target=worker_thread)
        thread.daemon = True
        thread.start()
        threads.append(thread)

    # Wait for all tasks in the queue to be processed.
    batch_queue.join()

    # Once all tasks are done, print the final results.
    print("\n-- All batches processed. Finalizing data. ---")
    print(f"Total recipes processed and categorized: {len(updated_recipes)}")

if __name__ == "__main__":
    main()
```
