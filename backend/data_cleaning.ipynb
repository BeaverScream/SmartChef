```python
import numpy as np

# Add additional column to distinguish meal, side-dishes, and desserts
df_with_meal_types = df_with_no_empty_tags.copy()

condition = [
    df_with_meal_types['tags'].apply(lambda x: any(tag in x for tag in ['desserts', 'pies-and-tarts', 'cakes', 'cookies-and-brownies', 'chocolates', 'drop-cookies'])),
    df_with_meal_types['tags'].apply(lambda x: any(tag in x for tag in ['side-dishes', 'appetizers', 'condiments-etc', 'dips', 'sauces']))
]

options = [
    'dessert',
    'side-dish'
]

df_with_meal_types['meal_type'] = np.select(condition, options, default='meal')

print(len(df_with_meal_types))
df_with_meal_types.head(30)
```

```python
# Cleaning tags as the data set now has a separate column to categorize the meal type

additional_tags_to_remove = [
    'side-dish',
    'main-dish',
    'dessert',
    'cookie',
    'pie',
    'tart',
    'cake',
    'chocolate',
    'appetizer',
    'condiment',
    'dip',
    'sauce',
    'beverage',
    'cocktail'
]

df_additional_tag_removed = df_with_meal_types.copy()

df_additional_tag_removed['tags'] = df_additional_tag_removed['tags'].apply(
    lambda tag_list: [
        tag for tag in tag_list 
        if tag.strip() and not any(word in tag for word in additional_tags_to_remove)
    ]
)

df_additional_tag_removed = df_additional_tag_removed[df_additional_tag_removed['tags'].str.len() > 0]

df_additional_tag_removed.head(20)
```

```python
# Add one more column 'food type' to distinguish dietary restriction
meat_keywords = ['meat', 'chicken', 'beef', 'pork', 'lamb', 'turkey', 'sausage', 'bacon', 'ham']
seafood_keywords = ['fish', 'tuna', 'salmon', 'shrimp', 'crab', 'squid', 'seafood', 'octopus']
vegan = ['vegan']
vegetarian = ['vegetarian']

conditions = [
    df_additional_tag_removed['tags'].apply(lambda x: any(tag in x for tag in meat_keywords)) | 
    df_additional_tag_removed['name'].str.lower().apply(lambda name: any(word in name for word in meat_keywords)),
    df_additional_tag_removed['tags'].apply(lambda x: (any(tag in x for tag in seafood_keywords)) & 
                                            (not any(tag in x for tag in meat_keywords))),
    df_additional_tag_removed['tags'].apply(lambda x: any(tag in x for tag in vegan)),
    df_additional_tag_removed['tags'].apply(lambda x: any(tag in x for tag in vegetarian))
]

options = [
    'meat-base',
    'pescetarian',
    'vegan',
    'vegetarian'
]

df_food_type_added = df_additional_tag_removed.copy()
df_food_type_added['food_type'] = np.select(conditions, options, default='uncategorized')

df_food_type_added.head(30)
```

```python
# One time operation to get generate embedding (all-MiniLM-L6-v2)
# Finally it worked!!!
import time
from sentence_transformers import SentenceTransformer

print("Loading the MiniLM model...")

# Getting the model
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully.")
print("-" * 50)

texts_to_embed = []
for _, row in df_sampled.iterrows():
    text = (
        f"Recipe Name: {row['name']}."
        f"Tags: {row['tags']}."
        f"Ingredients: {row['ingredients']}."
        f"Instructions: {row['steps']}."
        f"Description: {row['description']}."
        f"Food Type: {row['food_type']}."
    )
    texts_to_embed.append(text)

print(f"Generating embeddings for {len(texts_to_embed)} recipes...")
start_time = time.time()

# Generate all the embeddings in a single, efficient call.
# No batches, no API calls, no retries.
embeddings = model.encode(texts_to_embed)

end_time = time.time()

# Add the embeddings as a new column to the DataFrame.
df_sampled['embedding'] = embeddings.tolist()

print("-" * 50)
print("Embeddings generated successfully.")
print(f"Time taken: {end_time - start_time:.2f} seconds.")

df_sampled.head()
```

```python
# formatting recipes' steps (how to cook instructions)
import os
import sys
import pandas as pd
import ast

CSV_FILE = 'data/df_sampled_6000_MiniLM.csv'

try:
    df_sampled = pd.read_csv(CSV_FILE)
    print(f"Loaded {CSV_FILE}'.")
    
except FileNotFoundError:
    print(f"Error: CSV file '{CSV_FILE}' not found. Please ensure it is in the correct path.")
    sys.exit(1)
except Exception as e:
    print(f"Error loading the csv file: {e}")
    sys.exit(1)
    
df_sampled['name'] = df_sampled['name'].str.upper()

def format_instructions (steps):
    try:
        instructions = ast.literal_eval(steps)
    except Exception as e:
        return steps
    
    formatted_step = []
    for i, step in enumerate(instructions):
        step = step.strip()
        if not step:
            continue
        step = step[0].upper() + step[1:]
        formatted_step.append(f"{i+1}. {step}")
    return formatted_step

df_sampled['steps'] = df_sampled['steps'].apply(format_instructions)

df_sampled.head(6)

OUTPUT_CSV_FILE = 'data/df_sampled_6000_MiniLM.csv'
df_sampled.to_csv(OUTPUT_CSV_FILE, index=False)
print(f"Saved formatted CSV to '{OUTPUT_CSV_FILE}'")
```

```python
# Faiss index files generation

import numpy as np
import faiss
import os
import itertools 
import sys
import pandas as pd 
import ast 

# --- CONFIGURATION ---
BASE_DIR = 'data'
EMBEDDING_DIM = 384
DIETARY_PREFERENCES = ['vegan', 'vegetarian', 'meat-base', 'pescetarian']
CSV_FILE = 'data/df_sampled_6000_MiniLM.csv'

# --- LOAD AND PREPARE DATAFRAME ---
try:
    df_sampled = pd.read_csv(CSV_FILE)
    print(f"Loaded {len(df_sampled)} rows from '{CSV_FILE}'.")
    
    df_sampled['embedding'] = df_sampled['embedding'].apply(
        lambda x: np.array(ast.literal_eval(x), dtype=np.float32)
    )
    
except FileNotFoundError:
    print(f"Error: CSV file '{CSV_FILE}' not found. Please ensure it is in the correct path.")
    sys.exit(1)
except Exception as e:
    print(f"Error loading or processing embeddings from CSV: {e}")
    sys.exit(1)

# # --- VERIFY DATA FORMAT ---
# if 'food_type' in df_sampled.columns:
#     # Filter out empty/NaN values and take the first 5 for inspection
#     sample_food_types = df_sampled['food_type'].dropna().loc[lambda x: x != ''].head()
#     print("-" * 60)
#     print("Verifying 'food_type' column format (first 5 non-empty values):")
#     for i, val in sample_food_types.items():
#         print(f"  Index {i}: '{val}'")
#     print("-" * 60)
#     # Check for the correct data type
#     if not sample_food_types.empty and not isinstance(sample_food_types.iloc[0], str):
#          print("CRITICAL ERROR: 'food_type' column is not stored as strings. Please check CSV export format.")
#          sys.exit(1)
# else:
#     print("CRITICAL ERROR: 'food_type' column not found in DataFrame. Cannot proceed with filtering.")
#     sys.exit(1)
# # ---------------------------
    
# --- PRE-REQUISITE CHECKS ---
# This check ensures that the parsing above was successful and the column exists.
if 'embedding' not in df_sampled.columns:
    print('Error: The "embedding" column must be present and correctly generated in the DataFrame.')
    sys.exit(1)
    
print("-" * 60)
print(f"Starting generation of specialized Faiss indexes in '{BASE_DIR}'")
print(f"Total possible combinations: {2**len(DIETARY_PREFERENCES)}")
print("-" * 60)

# --- GENERATE COMBINATIONS OF FAISS INDEXES ---
total_combinations = 0
generated_files = []

# Generate all non-empty combinations (subsets) of the dietary preferences
# We start from 1 (single preference) up to len(DIETARY_PREFERENCES) (all preferences)
for r in range(1, len(DIETARY_PREFERENCES) + 1):
    for combination in itertools.combinations(DIETARY_PREFERENCES, r):
        total_combinations += 1
        pref_list = list(combination)
        
        # Create a unique, sorted key and file name for this combination
        pref_list.sort()
        file_key = "_".join(pref_list)
        file_path = os.path.join(BASE_DIR, f'faiss_{file_key}.bin')

        print(f"-> Processing combination: {file_key}...")
        
        # 1. FILTER DATAFRAME: Filter recipes that match ANY of the selected preferences
        filter_mask = df_sampled['food_type'].isin(pref_list)
        filtered_df = df_sampled[filter_mask]
        
        if filtered_df.empty:
            print(f"   [WARNING] No recipes found for this combination. Skipping index generation.")
            continue
            
        # 2. PREPARE VECTORS: Stack embeddings from the filtered DataFrame
        embedding_matrix = np.vstack(filtered_df['embedding'].tolist()).astype('float32')
        # Prepare IDs for mapping (Faiss requires a NumPy array of type int64)
        recipe_ids = filtered_df['id'].values.astype(np.int64)

        # --- FIXED FAISS WORKFLOW ---
        
        # 3a. CREATE BASE INDEX: IndexFlatL2 provides the search mechanism.
        # This must be empty before being wrapped. IndexIDMap doesn't accept faiss index that's not empty
        base_index = faiss.IndexFlatL2(EMBEDDING_DIM) 
        
        # 3b. CREATE ID MAP: Wrap the EMPTY base index with IndexIDMap.
        # This wrapper handles the translation from internal Faiss IDs to our database IDs.
        index_id_map = faiss.IndexIDMap(base_index)
        
        # 4. ADD DATA SIMULTANEOUSLY: Use add_with_ids to load vectors AND map IDs in one step.
        index_id_map.add_with_ids(embedding_matrix, recipe_ids)
        # --- END FIXED FAISS WORKFLOW ---

        # 5. SAVE INDEX FILE
        # We save the index_id_map, which contains both the base index and the ID mapping
        faiss.write_index(index_id_map, file_path)
        
        generated_files.append(file_path)
        print(f"   [SUCCESS] Index saved: {os.path.basename(file_path)} ({len(filtered_df)} vectors)")

print("-" * 60)
print(f"Indexing complete. Generated {len(generated_files)} specialized Faiss index files.")
```
